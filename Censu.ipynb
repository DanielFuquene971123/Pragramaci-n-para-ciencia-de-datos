# Detección de Valores Nulos

Para saber cuantos valores perdidos tenemos en cada una de las columnas (variables). Debemos usar la función .isnull().sum() la razon por la cual usamos en conjunto la función .sum() es por que va a estar sumando todos los valores que cumplan con la condicion null en cada una de las columnas.

```python
valores_null=censu.isnull().sum()
print(valores_null)
```

Como podemos observar en la variable 'workclass' tiene 963 con valores nulos (NaN). Tambien la variable 'occupation' con 966 valores nulos y la variable 'native-country' con 274 valores nulos. Para un total de 2.203 datos.

# Detección filas duplicadas

Aplicamos la función .duplicated() la cual nos dice mediante una confirmación con numeros booleanos si existe o no esta condición. De igual forma, adicionamos la función .sum() para determiar a lo largo de todas las filas la condición de filas duplicadas.

```python
filas_duplica=censu.duplicated().sum()
print(filas_duplica)
```

Como podemos observar ninguna fila tiene información duplicada en el dataframe de censu.

## Valores unicos
